```python
from huggingface_hub import notebook_login
from diffusers import StableDiffusionXLPipeline, AutoencoderKL
import torch
import os


notebook_login()
os.makedirs("outputs", exist_ok=True)
pipe = StableDiffusionXLPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    vae=AutoencoderKL.from_pretrained("madebyollin/sdxl-vae-fp16-fix", torch_dtype=torch.float16),
    torch_dtype=torch.float16,
    variant="fp16",
    use_safetensors=True
).to("cuda")

pipe.load_lora_weights("ThompsonBrbs24x/sdxl-base-1.0-pnemo-lora-v4", token=True)
pipe.to(torch.float16)

prompt = "An X ray of an human chest with pneumonia"

for i in range(10):
    image = pipe(prompt=prompt, num_inference_steps=40, guidance_scale=7.5).images[0]
    image_path = f"outputs/pneumonia_lora_image_{i+1}.png"
    image.save(image_path)
    print(f"✅ Image {i+1} enregistrée dans {image_path}")
```
